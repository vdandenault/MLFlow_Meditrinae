```
df = df.dropna(subset=['QS3', 'QS4A', 'QS4B', 'QS4C', 'QS4D'])
```

```
def cleanQS3(x):
    if x == 'Aucune douleur':
        return 'a'
    if x == 'Douleur qui vient et part':
        return 'b'
    if x ==  'Douleur toujours preÃÅsente':
        return 'c'
    else:
        print(x)
        return x
def cleanOUI_NON_TO_TRUE_FALSE(x):
    if x == 'Oui':
        return 'T'
    if x == 'Non':
        return 'F'
    else:
        print(x)
        return x
```

```
df_model = pd.DataFrame()
df_model['QS3'] = df['QS3']
df_model['QS3'] = df['QS3'].apply(cleanQS3)
df_model['QS3'].value_counts()
df['QS4A'].value_counts()
```

```
df_model['QS4A'] = df['QS4A'].apply(cleanOUI_NON_TO_TRUE_FALSE)
df_model['QS4B'] = df['QS4B'].apply(cleanOUI_NON_TO_TRUE_FALSE)
df_model['QS4C'] = df['QS4C'].apply(cleanOUI_NON_TO_TRUE_FALSE)
df_model['QS4D'] = df['QS4D'].apply(cleanOUI_NON_TO_TRUE_FALSE)
```

```
from pomegranate import *

QS4A = DiscreteDistribution({'T': 0.5, 'F': 0.5})
QS4B = DiscreteDistribution({'T': 0.5, 'F': 0.5})
QS4C = DiscreteDistribution({'T': 0.5, 'F': 0.5})
QS4D = DiscreteDistribution({'T': 0.5, 'F': 0.5})
QS4 = ConditionalProbabilityTable(
        [
        ['F', 'F', 'F', 'F', 'T', 0],
        ['F', 'F', 'F', 'T', 'T', 1],
        ['F', 'F', 'T', 'F', 'T', 1],
        ['F', 'F', 'T', 'T', 'T', 1],
        ['F', 'T', 'F', 'F', 'T', 1],
        ['F', 'T', 'F', 'T', 'T', 1],
        ['F', 'T', 'T', 'F', 'T', 1],
        ['F', 'T', 'T', 'T', 'T', 1],
        ['T', 'F', 'F', 'F', 'T', 1],
        ['T', 'F', 'F', 'T', 'T', 1],
        ['T', 'F', 'T', 'F', 'T', 1],
        ['T', 'F', 'T', 'T', 'T', 1],
        ['T', 'T', 'F', 'F', 'T', 1],
        ['T', 'T', 'F', 'T', 'T', 1],
        ['T', 'T', 'T', 'F', 'T', 1],
        ['T', 'T', 'T', 'T', 'T', 1],
        
        ['F', 'F', 'F', 'F', 'F', 1],
        ['F', 'F', 'F', 'T', 'F', 0],
        ['F', 'F', 'T', 'F', 'F', 0],
        ['F', 'F', 'T', 'T', 'F', 0],
        ['F', 'T', 'F', 'F', 'F', 0],
        ['F', 'T', 'F', 'T', 'F', 0],
        ['F', 'T', 'T', 'F', 'F', 0],
        ['F', 'T', 'T', 'T', 'F', 0],
        ['T', 'F', 'F', 'F', 'F', 0],
        ['T', 'F', 'F', 'T', 'F', 0],
        ['T', 'F', 'T', 'F', 'F', 0],
        ['T', 'F', 'T', 'T', 'F', 0],
        ['T', 'T', 'F', 'F', 'F', 0],
        ['T', 'T', 'F', 'T', 'F', 0],
        ['T', 'T', 'T', 'F', 'F', 0],
        ['T', 'T', 'T', 'T', 'F', 0]], [QS4A, QS4B, QS4C, QS4D])

s1 = State(QS4, name="QS4")
s2 = State(QS4A, name="QS4A")
s3 = State(QS4B, name="QS4B")
s4 = State(QS4C, name="QS4C")
s5 = State(QS4D, name="QS4D")
model = BayesianNetwork("QS4")
model.add_states(s2, s3, s4, s5, s1)
model.add_edge(s2, s1)
model.add_edge(s3, s1)
model.add_edge(s4, s1)
model.add_edge(s5, s1)
model.bake()

#print(model.predict_proba(['T', 'T', 'T', 'F', None]))
#print(model.plot())
```
